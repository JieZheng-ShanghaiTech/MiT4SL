```python
import pandas as pd
import numpy as np
```

```python
a375=pd.read_csv("/home/siyutao/SL/LukePi_2.0/cell_line/KG_Sequence/SL_data/MVGCNdata/A375/A375_GIscore_534.csv")
a375
```

```python
b=a375.drop_duplicates(subset=['gene1_primekg_index','gene2_primekg_index'],keep='last')
```

```python
b['label'].value_counts()
```

```python
b.reset_index(drop=True,inplace=True)
```

```python
b
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
df=b
# X 是特征，y 是标签
for i in range(5):
    sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42+i)
    X=df[['gene1_primekg_index','gene2_primekg_index','cell_id']].values
    y=df['label'].values
    for train_index, test_index in sss.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        X_train=pd.DataFrame(X_train)
        y_train=pd.DataFrame(y_train)
        df_x=pd.concat([X_train,y_train],axis=1)
        df_x.columns=[[0,1,2,3]]
        # df_x.to_csv(f"/home/siyutao/SL/LukePi_2.0/cell_line/KG_Sequence/SL_data/C1/sl_train_{i}.csv",index=False)
        df_x.to_csv(f"/home/siyutao/SL/LukePi_2.0/cell_line/KG_Sequence/SL_data/MVGCNdata/A375/C1_pure/sl_train_{i}.csv",index=False)
        X_test=pd.DataFrame(X_test)
        y_test=pd.DataFrame(y_test)
        dk_x=pd.concat([X_test,y_test],axis=1)
        dk_x.columns=[[0,1,2,3]]
        # dk_x.to_csv(f"/home/siyutao/SL/LukePi_2.0/cell_line/KG_Sequence/SL_data/C1/sl_test_{i}.csv",index=False)
        df_x.to_csv(f"/home/siyutao/SL/LukePi_2.0/cell_line/KG_Sequence/SL_data/MVGCNdata/A375/C1_pure/sl_test_{i}.csv",index=False)
```
